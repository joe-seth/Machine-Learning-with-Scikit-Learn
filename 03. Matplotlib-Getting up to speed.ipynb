### Two options for subplots

```python
# Subplot option 1
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,
                                            ncols=2,
                                            figsize=(10,5))
# Plot to each differet plot
ax1.plot(x, x/2)
ax2.scatter(np.random.random(10), np.random.random(10))
ax3.bar(nut_prices.keys(), nut_prices.values())
ax4.hist(np.random.randn(100));
```

```python
# Subplot option 2
fig, ax = plt.subplots(2,2, figsize=(10,5))
ax[0,0].plot(x, x/2)
ax[0,1].scatter(np.random.random(10), np.random.random(10))
ax[1,0].bar(nut_prices.keys(), nut_prices.values())
ax[1,1].hist(np.random.randn(100));
plt.show();
```

```python
# Let's try on another dataset
url = 'https://raw.githubusercontent.com/joe-seth/Machine-Learning-with-Scikit-Learn/main/datasets/heart-disease.csv'

heart_disease = pd.read_csv(url)
heart_disease
```

```python
# Create a histogram to view the distribution of the age column
heart_disease['age'].plot.hist(bins=50);
```

```python
heart_disease.plot.hist(figsize=(10,30),subplots=True);
```

```python
over_50 = heart_disease[heart_disease['age'] > 50 ]
over_50
```

```python
# Pyplot method
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target');
```

```python
# OO method
fig , ax = plt.subplots(figsize=(10,6))
over_50.plot(kind='scatter',
            x='age',
            y='chol',
            c='target',
            ax=ax);
ax.set_xlim([45,100])
```

```python
fig, ax = plt.subplots(figsize=(10,6))
# plot the data
scatter =  ax.scatter(x=over_50['age'],
                      y=over_50['chol'],
                      c=over_50['target']);
# Cusotmiz
ax.set(title='Heart Disease and Cholestral',
       xlabel='Age',
       ylabel='Cholestral')

ax.legend(*scatter.legend_elements(), title='Target');

# Add a horizontal line
ax.axhline(over_50['chol'].mean(),
          linestyle='--')
```

```python
# Subplot of chol age , thalach
fig, (ax0, ax1) = plt.subplots(2,1, figsize=(10,10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'])

# Customize ax0
ax0.set(title="Age vs Chol",
       xlabel="Age",
       ylabel="cholestral")
ax0.axhline(over_50['chol'].mean(),
           linestyle='--')
ax0.legend(*scatter.legend_elements(), title='Target');


# Add data to ax1
scatter2 = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'])
# Cutomize ax1
ax1.set(title="Age vs thalach",
       xlabel="Age",
       ylabel="Thalach")
ax1.axhline(over_50['thalach'].mean(),
           linestyle='--')
ax1.legend(*scatter.legend_elements(), title='Target');

# add title
fig.suptitle("Heart Diesease Analysis",fontsize=16, fontweight='bold');
```

## Customizing the x and  limitations

```python
# Subplot of chol age , thalach
fig, (ax0, ax1) = plt.subplots(2,1, figsize=(10,10), sharex=True)

# Add data to ax0
scatter = ax0.scatter(x=over_50['age'],
                     y=over_50['chol'],
                     c=over_50['target'],
                     cmap='winter')

# Customize ax0
ax0.set(title="Age vs Chol",
       xlabel="Age",
       ylabel="cholestral")
# Change the x axis limits
ax0.set_xlim([50,80])
ax0.axhline(over_50['chol'].mean(),
           linestyle='--')
ax0.legend(*scatter.legend_elements(), title='Target');


# Add data to ax1
scatter2 = ax1.scatter(x=over_50['age'],
                      y=over_50['thalach'],
                      c=over_50['target'],
                      cmap='winter')
# Cutomize ax1
ax1.set(title="Age vs thalach",
       xlabel="Age",
       ylabel="Thalach")
# Change the x axis limits
ax1.set_xlim([50,80])
ax1.set_ylim([60,200])
ax1.axhline(over_50['thalach'].mean(),
           linestyle='--')
ax1.legend(*scatter.legend_elements(), title='Target');

# add title
fig.suptitle("Heart Diesease Analysis",fontsize=16, fontweight='bold');
```
